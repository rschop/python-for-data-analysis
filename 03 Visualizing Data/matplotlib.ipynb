```python
import matplotlib.pyplot as plt
print(plt)
```

```python
import numpy as np
x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
plt.title('Sin from 0 to 2*pi')
plt.plot(x,y)
plt.show()
```

```python
plt.title('Sin from 0 to 2*pi')
plt.xlabel('Input')
plt.ylabel('Output')
plt.plot(x,y)
plt.show()
```

```python
y2 = np.cos(x)
plt.title('Sin and Cos from 0 to 2*pi')
plt.xlabel('Input')
plt.ylabel('Output')
plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```

```python
plt.title('Sin and Cos from 0 to 2*pi')
plt.xlabel('Input')
plt.ylabel('Output')
plt.plot(x, y, label='sine')
plt.plot(x, y2, label='cosine')
plt.legend()
plt.show()
```

```python
plt.title('Minus Sin and Minus Cos from 0 to 2*pi')
plt.xlabel('Input')
plt.ylabel('Output')
plt.plot(x, -y, label='sine')
plt.plot(x, -y2, label='cosine')
plt.legend()
plt.show()
```

```python
plt.title('Sin and Cos from 0 to 2*pi')
plt.xlabel('Input')
plt.ylabel('Output')
plt.plot(x, -y, label='sine')
plt.plot(x, -y2, label='cosine')
plt.legend(loc='lower center')
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.title('Sin and Cos from 0 to 2*pi')
plt.ylabel('Sin')
plt.plot(x, y)

plt.subplot(2, 1, 2)
plt.ylabel('Cos')
plt.plot(x, y2)
plt.show()
```

**Markers** -  https://matplotlib.org/api/markers_api.html#module-matplotlib.markers

**Colors** - https://matplotlib.org/api/_as_gen/matplotlib.pyplot.colors.html

```python
plt.title('Minus Sin and Minus Cos from 0 to 2*pi')
plt.xlabel('Input')
plt.ylabel('Output')
plt.plot(x, -y, 'ro', label='sine')
plt.plot(x, -y2, label='cosine')
plt.legend()
plt.show()
```

```python
plt.title('Minus Sin and Minus Cos from 0 to 2*pi')
plt.xlabel('Input')
plt.ylabel('Output')
plt.plot(x, -y, 'ro', label='sine')
plt.plot(x, -y2, marker='^', color='g', label='cosine', linewidth=0)
plt.legend()
plt.show()
```
